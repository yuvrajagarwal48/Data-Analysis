# Coffee Analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
coffee_df=pd.read_csv("df_arabica_clean.csv")
```

```python
coffee_df
```

```python
coffee_df.info()
```

```python
coffee_df.describe()
```

```python
features=["ID","Country of Origin","Company","Altitude","Region","Variety","Processing Method","Aroma","Flavor","Aftertaste","Acidity","Body","Balance","Moisture Percentage","Color"]
coffee=coffee_df[features]
```

```python
coffee
```

```python
print(coffee["Company"].nunique())
print(coffee["Variety"].nunique())
print(coffee["Processing Method"].nunique())
```

Giving rating to each coffee by averaging over the ratings of Aroma,Flavor,Aftertaste,Acidity,Body,Balance,Uniformity,Sweetness,Overall

```python
coffee['rating']=coffee[["Aroma","Flavor","Aftertaste","Acidity","Body","Balance"]].mean(axis=1)
```

```python
coffee.sample(10)
```

## Figuring out which countries have most Coffee Companies

```python
country=coffee["Country of Origin"].value_counts()
country
```

```python
plt.figure(figsize=(12,6))
plt.xticks(rotation=75)
plt.title("Top Countries")
sns.barplot(x=country.index,y=country)
```

## Seeing which company has the most manufacturing of coffee

```python
top_companies=coffee['Company'].value_counts()
top_companies
```

```python
plt.figure(figsize=(15,6))
plt.xticks(rotation=90)
plt.title("Top Companies")
sns.barplot(x=top_companies.index,y=top_companies);
```

### Keeping only upper limit of altitude and converting all to int

```python
altitudes=coffee.Altitude.astype(str).map(lambda x: x.split('-')[-1])
altitudes.value_counts().sort_index()
```

```python
plt.figure(figsize=(15,6))
plt.xticks(rotation=90)
plt.plot(altitudes.value_counts().sort_index())
```

#### Hence most coffee's are grown at altitude 1200

## Most used processing methods

```python
methods=coffee['Processing Method'].value_counts()
methods
```

```python
plt.rcParams["figure.figsize"] = [12, 6]
labels = methods.index
patches, text= plt.pie(methods, startangle=90)
plt.legend(patches, labels, loc="best")
plt.axis('equal')
```
